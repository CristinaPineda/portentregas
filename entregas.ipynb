```python
import pandas as pd
import numpy as np
import datetime
import matplotlib
import matplotlib.pyplot as plt
```

```python
entregas = pd.read_excel('entregas.xlsx')
entregas
```

```python
entregas1 = entregas['Sprint'] > 0
entregasOn = entregas[entregas1]
entregasOn
```

```python
#substitue nulos 
entregasOn.fillna(value = 0, inplace = True)
entregasOn
```

```python
entregasOn1 = entregasOn[['Produto', 'Atividades', 'total']]
entregasOn1 = entregasOn1.replace({"Consignado CL 26": "Consignado CL"})
entregasOn1 = entregasOn1.replace({"Consignado CL 27": "Consignado CL"})
entregasOn1 = entregasOn1.replace({"Consignado OP 01": "Consignado OP"})
entregasOn1 = entregasOn1.replace({"Consignado OP 02": "Consignado OP"})
entregasOn1 = entregasOn1.replace({"Consignado OP 03": "Consignado OP"})
entregasOn1
```

```python
produtos = {'produto': ['Giro California','Consignado Cl', 'Consignado OP'], 'total' : [1.0, 37.39, 3.10]}
produtos1 = pd.DataFrame(produtos)
produtos1
```

## Entrega por produto:

```python
x= ['Consignado Cl', 'Consignado OP','Giro California']
y= [37.39, 3.10,1.0]
plt.figure(figsize=(20, 10))
plt.barh(x, y, color=('darkcyan', 'orangered', 'orange'))
plt.xlabel('Horas',fontsize=15)
plt.ylabel('Produtos',fontsize=15)
plt.title('Entrega por produtos', fontsize=20)
plt.xticks(y, fontsize=18)
plt.show()
```

```python
consignadoCL = entregasOn1['Produto'] == 'Consignado CL'
filtered_df = entregasOn1[consignadoCL]
filtered_df.style.format({'total':'{}h'})
```

```python
x= filtered_df['Atividades']
y= [9.41, 4.30,7.30, 2.45,10.28,2.45]
plt.figure(figsize=(28, 10))
plt.bar(x, y, color=('blue', 'red','red', 'blue', 'orangered', 'cyan'))
plt.xlabel('Atividades',fontsize=18)
plt.ylabel('horas',fontsize=15)
plt.title('Atividades Consignado CL', fontsize=20)
# plt.xticks(y, fontsize=14)

def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i, y[i], y[i], ha = 'center', fontsize=14)
addlabels(x, y)
plt.show()
```
