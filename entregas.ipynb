```python
import pandas as pd
import numpy as np
from datetime import datetime
import time
import matplotlib.pyplot as plt
```

### Leitura de planilha de entregas

```python
entregas = pd.read_excel('entregas2.xlsx')
entregas
```

### Sprint analisada

Atribua à variavel *sprint* o número da sprint a ser analisada

```python
sprint = 5
entregas_sprint = entregas['Sprint'] == sprint
entrega_sprint_horas = entregas[entregas_sprint]
#substitue nulos 
entrega_sprint_horas.fillna(value = '00:00:00', inplace = True)
entrega_sprint_horas
```

## Detalhado por atividade

```python
entregasOn1 = entrega_sprint_horas[['Produto', 'Atividades', 'total']]
entregasOn1.sort_values(['Produto','Atividades'],ascending=True)
## Por produto
```

# Entrega detalhada por produto

## Consignado

```python
consignado = entregasOn1['Produto'] == 'Consignado'
filtered_df = entregasOn1[consignado]
#filtered_df.style.format({'total':'{}h'})
filtered_df
```

```python
x=('Repositório de saldo',
'Reunião Alinhamento',
'Analise POF 26',
'Integrado em HML30',
'Integrado em HML31',
'Analise POF 30/31',
'Integrado em HML27',
'Inception Repac',
'Apoio- Tombamento TTs',
'Integrado em HML26')
y= (1.0, 3.35, 2.0, 5.08, 5.10, 2.40, 3.30, 1.30, 3.29, 1.30)
plt.figure(figsize=(30, 5))
plt.bar(x, y, color=('blue', 'red','red', 'blue', 'orangered', 'cyan','blue', 'red', 'blue', 'orangered', 'cyan'))
plt.xlabel('Atividades',fontsize=12)
plt.ylabel('horas',fontsize=15)
plt.title('Atividades Consignado', fontsize=20)
plt.xticks(x)

def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i, y[i], y[i], ha = 'center', fontsize=12)
        
addlabels(x, y)
plt.show()
```
